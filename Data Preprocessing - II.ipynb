```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.mode.chained_assignment = None 
from scipy.stats import skew
```

```python
X_train = pd.read_csv(r'C:\Users\ArpitKh4101\Downloads\house-prices-advanced-regression-techniques/a.csv')
X_test = pd.read_csv(r'C:\Users\ArpitKh4101\Downloads\house-prices-advanced-regression-techniques/b.csv')
```

```python
data = pd.concat([X_train,X_test])
data.head()
```

```python
# Feature Engineering

data['TotalArea'] = data['TotalBsmtSF'] + data['1stFlrSF'] + data['2ndFlrSF'] + data['GrLivArea'] +data['GarageArea']
```

```python
data['Bathrooms'] = data['FullBath'] + data['HalfBath']*0.5 
data = data.drop({'FullBath','HalfBath'},axis=1)
data['BsmtBathrooms'] = data['BsmtFullBath'] + data['BsmtHalfBath']*0.5 
data = data.drop({'BsmtFullBath','BsmtHalfBath'},axis=1)
data['BHKAbvGr'] = data['KitchenAbvGr'] + data['BedroomAbvGr']
```

```python
data.shape
```

```python
X_train = data[:1460]
X_test = X_test[1460:]
```

```python
X_train.to_csv(r'C:\Users\ArpitKh4101\Downloads\house-prices-advanced-regression-techniques/FE_train.csv',index= False)
X_test.to_csv(r'C:\Users\ArpitKh4101\Downloads\house-prices-advanced-regression-techniques/FE_test.csv',index= False)
```

```python
num = data.dtypes[data.dtypes != "object"].index
skewed = X_train[num].apply(lambda x: skew(x.dropna()))
```

```python
skewed = skewed[skewed > 0.75]
```

```python
skewed = skewed.index
X_train[skewed] = np.log1p(X_train[skewed])
```
